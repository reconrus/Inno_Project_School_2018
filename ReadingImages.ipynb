#  Использование OpenCV

Считывание, отображение и запись изображений при помощи  методов **`cv2.imread()`**,**`cv2.imshow()`**, **`cv2.imwrite()`**
___

## Чтение изображений
---
Используйте функцию **`cv2.imread()`**, чтобы считать изображение. Изображение должно находиться в той же папке, либо должен быть указан полный путь к ней.

Второй аргумент это булевский флаг, который говорит как должно быть прочитано изображение:

- **`cv2.IMREAD_COLOR`** : Стандартный флаг, считывает изображение в стандартных цветах
- **`cv2.IMREAD_GRAYSCALE`** :Считывает изображение в серых тонах
- **`cv2.IMREAD_UNCHANGED`** : Считывает изображение как таковое, включая альфа канал

Вместо этих трех флагов вы можете просто использовать **1, 0, -1**

___

```python
import numpy as np
import cv2

# Загружает изображение в серых тонах
img = cv2.imread('img/jobscol.jpg',0)
```

## Вывод изображений
---
Для вывода изображений используется библиотека `matplotlib` и функция **`matplotlib.pyplot.imshow`**
___

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img, cmap = 'gray')
plt.xticks([]), plt.yticks([])  # для того чтобы скрыть координатные оси
plt.show()
```

```python
img2 = cv2.imread('img/jobscol.jpg',1)
plt.imshow(img2)
plt.xticks([]), plt.yticks([])  # для того чтобы скрыть координатные оси
plt.show()
```

```python
img3 = img2[:,:,::-1]
plt.imshow(img3)
plt.xticks([]), plt.yticks([])  # для того чтобы скрыть координатные оси
plt.show()
```

## Сохранение изображений
---
Для сохрания изображений используется `cv2.imwrite()`
___

```python
cv2.imwrite('jobsgray.png',img)
```

# Функционал для рисования объектов
---
В этом разделе вы научитесь рисовать различные геометрические объекты и рассмотрите следующие методы для этого **` cv2.line(), cv2.circle() , cv2.rectangle(), cv2.ellipse(), cv2.putText()`** и т.д.
___

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Линия
---
Чтобы нарисовать линию, вам нужно задать начальную точку линии и конечную. Мы создадим черную картинку и нарисуем синию линию от верхнего левого угла до правого нижнего
___

```python
# Создаем черный фон
img = np.zeros((512,512,3), np.uint8)
plt.imshow(img)
```

```python
# Синяя диагональная линия с толщиной 5 пикс.
img = cv2.line(img,(0,0),(512,512),(0,0,255),5)
plt.imshow(img)
```

### Прямоугольник
---
Чтобы нарисовать прямоугольник, вам нужно задать координаты левого верхнего и правого нижнего углов прямоугольника. На этот раз мы нарисуем зеленый прямоугольник в верхнем правом углу
___

```python
img = cv2.rectangle(img,(384,3),(510,128),(0,255,0),3)
plt.imshow(img)
```

### Круг
---
Чтобы нарисовать круг, вам нужно задать его центр и радиус. Мы нарисуем круг внутри квадрата
___

```python
img = cv2.circle(img,(447,63), 60, (255,0,0), -1)
plt.imshow(img)
```

### Эллипс
---
Чтобы нарисовать эллипс нам нужно задать несколько аргументов. Первый аргумент это центр нашего эллипса. Следующий аргумент это длина осей (длина мажорной оси, длина минорной оси). **`Угол`** это угол вращения эллипса в направление против часовой стрелки.**`Стартовый угол`** и **`Конечный угол`** обозначают начало и конец дуги эллипса в направлении противоположном часовой стрелки от мажорной оси. Мы нарисуем половину эллипса в центре картинки.
___

```python
img = cv2.ellipse(img,(256,256),(100,50),0,0,180,(0,0,255),-1)
plt.imshow(img)
```

### Полигон
---
Чтобы нарисовать полигон для начала вам нужно задать его вершины. Задайте эти точки массивом размерностью `ВЕРШИНЫх1х2`. 
___

```python
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
img = cv2.polylines(img,[pts],True,(255,255,0))
plt.imshow(img)
```
