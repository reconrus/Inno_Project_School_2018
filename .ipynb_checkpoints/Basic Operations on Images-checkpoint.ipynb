## Доступ и изменение значений пикселя


Для начала загрузим цветное изображение

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('img/jobscol.jpg')
plt.imshow(img)
```

---
**Изображение представленно в виде двумерного массива, поэтому можно получить доступ к значению пикселя, как к ячейки массива**
___

```python
px = img[100,100]
print(px)
```

```python
# Доступ только к голубому пикселю
blue = img[100,100,0]
print(blue)
```

```python
# Изменение значения пикселя
img[100,100] = [255,255,255]
print(img[100,100])
plt.imshow(img)
```

```python
# Параметры изображения
print(img.shape)
# Размер изображения
print(img.size)
# Тип данных изображения
print(img.dtype)
```

## Регион изображения (Image ROI)

```python
img = cv2.imread('img/jobscol.jpg')
face = img[0:250, 150:350]
img[0:250, 400:600] = face
plt.imshow(img)
```

## Разбиение и слияния каналов изображения

```python
img = cv2.imread('img/jobscol.jpg')
b,g,r = cv2.split(img)
img = cv2.merge((r,g,b))
plt.imshow(img)
```

## Смешивание изображений
---
Смешивание изображений, это тоже самое, что и арифметическое сложение чисел. Каждое изображение имеет вес. При смешивании берутся различные веса от каждого элемента и вычисляется новое изображение по следующей форме:

$dst=\alpha*f_0(x)+\beta*f_1(x)+\gamma$
___

```python
img1 = cv2.imread('img/van_gogh.jpg')
img2 = cv2.imread('img/eiffel.jpg')

dst = cv2.addWeighted(img1,0.3,img2,0.7,0)

plt.imshow(dst)
```

## Битовые операции

```python
img1 = cv2.imread('img/jobscol.jpg')
img2 = cv2.imread('img/apple.png')

b,g,r = cv2.split(img1)
img1 = cv2.merge((r,g,b))

# Поставим изображение в левый верхний угол, для этого создадим ROI
rows, cols, channels = img2.shape
roi = img[0:rows,0:cols]

# Создадим маску логотипа и создадим ее инверснутую версию
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)

# Затемним регион логотипа в ROI
img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv)

# Возьмем только регион изображения из изображения логотипа
img2_fg = cv2.bitwise_and(img2,img2,mask = mask)

# Размещаем наше изображение
dst = cv2.add(img1_bg,img2_fg)
img1[0:rows, 0:cols ] = dst
plt.imshow(img1)
```
